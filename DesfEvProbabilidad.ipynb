Desafío 1: Generación de funciones

● Genere funciones para calcular la media y varianza de un vector. Debe cumplir con los siguientes requisitos:
○ Ambas funciones deben ingresar un argumento x correspondiente al vector.
○ Las funciones deben contener docstrings con la documentación asociada a la variable y retornar un valor utilizando return.
○ La función de la varianza se debe llamar a la función de la media.
● Utilice las funciones para reportar la información sobre goles_favor, goles_contra y puntos.

```python
def calculo_medidas(x):
    tmp_mean= np.mean(x)
    tmp_var= np.var(x)
    
    return np.round(tmp_mean,2),np.round(tmp_var,2)
```

```python
calculo_medidas(df['goles_favor']) #Esto devuelve una tupla
```

```python
calculo_medidas(df['goles_contra'])
```

```python
calculo_medidas(df['puntos'])
```

Desafío 4: Función simuladora
● Genere una función llamada simulate_pets_prob que tome como argumento un
número finito de simulaciones a generar.
● La función debe simular dos situaciones young_pet y old_pet, y contar la
ocurrencia de los siguientes casos:
○ De las dos mascotas simuladas (young y old), contar las ocasiones donde
por lo menos una de las mascotas sea un perro.
○ De las dos mascotas simuladas, contar las ocasiones donde old_pet sea un
perro.
○ De las dos mascotas simuladas, contar las ocasiones donde los dos sean
perros.
● El método debe tener una semilla pseudoaleatoria de 1.
● El output de la función debe entregarse en términos de Probabilidad.
● De los tres escenarios, ¿Cuál es el menos probable? ¿Cuál es el más probable? ¿Por
qué?

```python
def generate_pet():
    """docstring for generate_pet"""
    return np.random.choice(['Gato', 'Perro'])
generate_pet()

```

```python

def simulate_pets_pr(n_sims):
    """docstring for simulate_pets_pr"""
    dos_perros = 0
    perro_viejo = 0
    at_least_un_perro = 0
    #aquí est ala semilla fijada en 1
    np.random.seed(1)
    for _ in range(n_sims):
        young_pet = generate_pet()
        old_pet = generate_pet()
        if old_pet == 'Perro':
            perro_viejo += 1
        if old_pet == 'Perro' and young_pet == 'Perro':
            dos_perros += 1
        if old_pet == 'Perro' or young_pet == 'Perro':
            at_least_un_perro += 1

    #De las dos mascotas simuladas, contar las ocasiones donde los dos sean perros.
    print("Pr(2 perros): ",dos_perros/n_sims)
    # De las dos mascotas simuladas, contar las ocasiones donde old_pet sea un perro.
    print("Pr(2 perros | perro viejo): ", dos_perros / perro_viejo)
    # De las dos mascotas simuladas (young y old), contar las ocasiones donde por lo menos una de las mascotas sea un perro.
    print("Pr(2 perros | por lo menos 1 perro): ", dos_perros / at_least_un_perro)
simulate_pets_pr(10)
```

De los tres escenarios, 
¿Cuál es el más probable?  Pr(2 perros|perro viejo)
¿Cuál es el menos probable?   Pr(2 perros)
